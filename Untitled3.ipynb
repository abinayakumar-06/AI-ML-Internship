```python
#trimming and removing whitespace
n="     abinaya   "
print(n.strip())
print(n.lstrip())
print(n.rstrip())
```

```python
#testing string contents
n="abinayakumar"
print(n.isalpha())
m="ps123"
print(m.isalnum())
p="123"
print(p.isdigit())
s="abinaya"
print(s.islower())
print(s.isupper())
a="ice cream"
print(a.istitle())
f="ice"
print(f.isspace())
```

```python
#set
s=set()
print(type(s))
s={1,2,3,4,4,5,6}
print(s)
s.clear()
print(s)
print(s.copy())
p={4,5,7,8,9}
print(p.difference(s))
print(p.difference_update(s))
print(p)
print(p.intersection(s))
p.intersection_update(s)
print(p)
s.discard(5)
print(s)
print(s.isdisjoint(p))
print(s.symmetric_difference(p))
s.symmetric_difference_update(p)
print(s)
x={1,2,3,4}
y={2,3}
print(x.issuperset(y))
print(y.issubset(x))
y.remove(2)
print(y)
print(y.union(x))
y.update({44,55,66})
print(y)
```
