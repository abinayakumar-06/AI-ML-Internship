```python
#pattern
for i in range(5):
  for j in range(5):
    print("*",end="")
  print()
```

```python
for i in range(1,6):
  for j in range(i):
    print("*",end="")
  print()
```

```python
n=5
for i in range(1,n+1):
  for j in range(1,n+1):
    if (j==i or i==n or j==1):
      print("*",end="")
    else:
      print(" ",end="")
  print()
```

```python
#while loop increment
i=1
while i<=10:
  print(i)
  i+=1
```

```python
#while loop decrement
i=10
while i>=1:
  print(i)
  i-=1
```

```python
#jumping/transfer control
name=["abi","anu","hema","hari"]
for i in name:
  if i=="hari":
    break
  print(i)
```

```python
name=["abi","anu","hema","hari"]
for i in name:
  if i=="anu":
    continue
  print(i)
```

```python
#number square
for i in range(1,6):
  for j in range(5):
    print(i,end="")
  print()
```

```python
#reverse triangle
for i in range(5,0,-1):
  for j in range(i):
    print("*",end="")
  print()
```

```python
#hollow square
n=5
for i in range(n):
  for j in range(n):
    if i==0 or i==n-1 or j==0 or j==n-1:
      print("*",end="")
    else:
      print(" ",end="")
  print()
```

```python
#X pattern
for i in range(5):
  for j in range(5):
    if i==j or i+j==n-1:
      print("*",end="")
    else:
      print(" ",end="")
  print()
```
